```python
data=pd.read_csv('turnstyle_info.csv')
data.columns=data.columns.str.strip()
data.head()
```

```python
#not necesary
monday=data[data.DATE=='6/12/17']
monday=monday.reset_index(drop=True)
monday.head()
```

```python
#not necesary
monday_dict={}
for i in range(len(monday.index)):
    tup=(monday['C/A'][i],monday['UNIT'][i],monday['SCP'][i],monday['STATION'][i])
    if tup not in monday_dict:
        monday_dict[tup]=[(monday['TIME'][i],monday['EXITS'][i])]
    else:
        monday_dict[tup].append((monday['TIME'][i],monday['EXITS'][i]))
print(monday_dict)
```

```python
daily_dict={}
for i in range(len(data.index)):
    tup=(data['C/A'][i],data['UNIT'][i],data['SCP'][i],data['STATION'][i])
    if data['DATE'][i] not in daily_dict:
        daily_dict[data['DATE'][i]]={}
    if tup not in daily_dict[data['DATE'][i]]:
        daily_dict[data['DATE'][i]][tup]=[(data['TIME'][i],data[data.columns[-1]][i])]
    else:
        daily_dict[data['DATE'][i]][tup].append((data['TIME'][i],data[data.columns[-1]][i]))
```

```python
print(daily_dict['6/12/17'])
```

```python
#{date:{(station/turnstile):total exits}}
total_turnstile={}
for date,values in daily_dict.items():
    for stations_turns,times_exits in values.items():
        if date not in total_turnstile.keys():
            total_turnstile[date]={stations_turns:times_exits[-1][1]-times_exits[0][1]}
        else:
            total_turnstile[date][stations_turns]=times_exits[-1][1]-times_exits[0][1]
print(total_turnstile['6/12/17'])
```

```python
total_station={}
for date,values in total_turnstile.items():
    for info,riders in values.items():
        tup=(info[0],info[1],info[3])
        if date not in total_station.keys():
            total_station[date]={tup:riders}
        if tup not in total_station[date].keys():
            total_station[date][tup]=riders
        else:
            total_station[date][tup]+=riders
print(total_station['6/15/17'])
```

```python
weekday_mornings=pd.read_csv('turnstyle_info.csv')
weekday_mornings.columns=weekday_mornings.columns.str.strip()
weekday_mornings.head()
```

```python
weekday_mornings['HOUR']=pd.to_datetime(weekday_mornings['TIME']).dt.hour
weekday_mornings.head()
```

```python
weekday_mornings['DAY']=pd.to_datetime(weekday_mornings['DATE']).dt.day
weekday_mornings.head()
```

```python
weekday_mornings.info()
```

```python
weekday_mornings_2a=weekday_mornings[(weekday_mornings['DAY']==12)]
weekday_mornings_2=weekday_mornings_2a[((weekday_mornings['HOUR']>=5)&(weekday_mornings['HOUR']<12))|((weekday_mornings['HOUR']==12)&(weekday_mornings['MINUTE']==0))]
weekday_mornings_2=weekday_mornings_2.reset_index(drop=True)
weekday_mornings_2.head()
```

```python
weekday_mornings_3=weekday_mornings_2.sort_values(by=['DAY','STATION','SCP','UNIT','HOUR'])
weekday_mornings_3=weekday_mornings_3.reset_index(drop=True)
weekday_mornings_3.head()
```

```python
wm3={}
for i in range(len(weekday_mornings_3)):
    if weekday_mornings_3['STATION'][i] not in wm3.keys():
        wm3[weekday_mornings_3['STATION'][i]]={}
    if (weekday_mornings_3['C/A'][i],weekday_mornings_3['UNIT'][i],weekday_mornings_3['SCP'][i]) not in wm3[weekday_mornings_3['STATION'][i]].keys():
        wm3[weekday_mornings_3['STATION'][i]][(weekday_mornings_3['C/A'][i],weekday_mornings_3['UNIT'][i],weekday_mornings_3['SCP'][i])]=[weekday_mornings_3['EXITS'][i]]
    else:
        wm3[weekday_mornings_3['STATION'][i]][(weekday_mornings_3['C/A'][i],weekday_mornings_3['UNIT'][i],weekday_mornings_3['SCP'][i])].append(weekday_mornings_3['EXITS'][i])
print(wm3)
```

```python
for station,values in wm3.items():
    for scp,exit_lst in values.items():
        if len(exit_lst)==1:
            print(station,exit_lst)
```

```python
wm4={}
for station,values in wm3.items():
    for scp,exit_lst in values.items():
        if len(exit_lst)==1:
            continue
        if station not in wm4.keys():
            wm4[station]=[exit_lst[-1]-exit_lst[0]]
        else:
            wm4[station].append(exit_lst[-1]-exit_lst[0])
print(wm4)
```

```python
wm5={}
for station,exit_lst in wm4.items():
    wm5[station]=sum(exit_lst)
print(sorted(wm5.items(),key=lambda x:x[1],reverse=True))
```
